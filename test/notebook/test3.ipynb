```python
# SEC에서 제공하는 XBRL 프레임 데이터 API를 호출하여 AccountsPayableCurrent 데이터를 가져옵니다.
xbrl_url = "https://data.sec.gov/api/xbrl/frames/us-gaap/AccountsPayableCurrent/USD/CY2019Q1I.json"
xbrl_headers = {
    "User-Agent": "jwlee <rgm418@gmail.com>"
}
xbrl_response = requests.get(url=xbrl_url, headers=xbrl_headers)
xbrl_data = xbrl_response.json()
```

```python
# xbrl_data의 'data' 키에 있는 리스트를 DataFrame으로 변환합니다.
xbrl_df = pd.DataFrame(xbrl_data['data'])
```

```python
# xbrl_df에서 CIK 값을 하나 선택하여 SEC 제출 정보 API에서 데이터를 가져옵니다.
# 예시로 첫 번째 행의 CIK를 사용합니다.
sample_cik = str(xbrl_df.iloc[0]['cik']).zfill(10)
submission_url = f"https://data.sec.gov/submissions/CIK{sample_cik}.json"
submission_headers = {
    "User-Agent": "jwlee <rgm418@gmail.com>"
}
submission_response = requests.get(url=submission_url, headers=submission_headers)
submission_data = submission_response.json()

# 결과 확인 (예: 최근 제출 문서 리스트 출력)
recent_filings = submission_data.get('filings', {}).get('recent', {})
print("최근 제출 문서:", recent_filings)
```
