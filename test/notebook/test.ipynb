```python
url = "https://opendart.fss.or.kr/api/elestock.json"
params = {
    "crtfc_key": api_key,
    "corp_code": "00126380"
}

response = requests.get(url, params=params)
if response.status_code == 200:
    print(response.json())
else:
    print(f"요청 실패: {response.status_code}")
```

```python
elestock_data = response.json()
elestock_df = pd.DataFrame(elestock_data['list'])
elestock_df.insert(0, 'status', elestock_data['status'])
elestock_df.insert(1, 'message', elestock_data['message'])
```
